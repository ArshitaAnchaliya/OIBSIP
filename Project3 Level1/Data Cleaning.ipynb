#### Outlier Detection

```python
#taking price column

mean_price = datafile['price'].mean()
median_price = datafile['price'].median()
std_price = datafile['price'].std()
Q1 = datafile['price'].quantile(0.25)
Q3 = datafile['price'].quantile(0.75)
IQR = Q3 - Q1
print(f"Mean: {mean_price}, Median: {median_price}, Standard Deviation: {std_price}")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
```

```python
# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
# Identify outliers
outliers = datafile[(datafile['price'] < lower_bound) | (datafile['price'] > upper_bound)]
print("Identified Outliers:")
print(outliers['price'])
```

```python
# Get distinct (unique) values of 'price' from outliers
distinct_outlier_prices = outliers['price'].unique()

# Print distinct outlier prices
print("Distinct Outlier Prices:")
print(distinct_outlier_prices)
```
